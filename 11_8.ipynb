General motion of a two cart, three spring system
$$
\begin{bmatrix} 
1\\
1
\end{bmatrix}
(B_1\cos(\omega_1 t) + B_2\sin(\omega_1 t))
+
\begin{bmatrix}
1\\ -1
\end{bmatrix}
(C_1\cos(\omega_2 t) + C_2\sin(\omega_2 t))
$$
Where $\omega_1 = \sqrt\frac{k}{m}$ and $\omega_2 = \sqrt\frac{3k}{m}$

```python
#constant and initial conditions
#at t=0 cart speeds ate v_0 in opposite directions
omega_1 = 1
omega_2 = np.sqrt(3)
v_0 = 1
```

```python
def x_1_b(t):
  return (-omega_1/omega_2)*np.sin(omega_2*t)

def x_2_b(t):
  return (omega_1/omega_2)*np.sin(omega_2*t)
```

```python
t = np.linspace(0,30,1001)
x1_b = x_1_b(t)
x2_b = x_2_b(t)
```

```python
fig1, ax1 = plt.subplots(2)
ax1[0].plot(t,x1_b)
ax1[1].plot(t,x2_b)

ax1[0].set_xlabel('Time (s)')
ax1[1].set_xlabel('Time (s)')
ax1[0].set_ylabel('x(t)')
ax1[1].set_ylabel('x(t)');
```

We see very simple oscilltion here due to only one term in one normal mode being active. 

```python
def x_1_c(t):
  return np.sin(omega_1*t) + (-omega_1/omega_2)*np.sin(omega_2*t)

def x_2_c(t):
  return np.sin(omega_1*t) + (omega_1/omega_2)*np.sin(omega_2*t)
```

```python
x1_c = x_1_c(t)
x2_c = x_2_c(t)
```

```python
fig2, ax2 = plt.subplots(2)
ax2[0].plot(t,x1_c)
ax2[1].plot(t,x2_c)

ax2[0].set_xlabel('Time (s)')
ax2[1].set_xlabel('Time (s)')
ax2[0].set_ylabel('x(t)')
ax2[1].set_ylabel('x(t)');
```

The motion is much more sporatic in this case because both normal modes have a term active.

```python
def sin_map(r,x):
  return r*sin(pi*x)
```

```python
def x_t(r):  
  x_new = [x_0]
  for i in range(x_len):
    x_new.append(sin_map(r,x_new[i]))

  return x_new
```

```python
r1 = 0.60
r2 = 0.79
r3 = 0.85
r4 = 0.865
x_len = 100
x_0 = 0.3
t = linspace(0,20,101)
```

```python
x1 = x_t(r1)
x2 = x_t(r2)
x3 = x_t(r3)
x4 = x_t(r4)
```

```python
fig,ax = plt.subplots(4, figsize = (6,16))
ax[0].plot(t,x1,'o',label='r = 0.60')
ax[0].set_ylabel('$x_t$')
ax[1].plot(t,x2,'o', label='r = 0.79')
ax[1].set_ylabel('$x_t$')
ax[2].plot(t,x3,'o', label = 'r = 0.85')
ax[2].set_ylabel('$x_t$')
ax[3].plot(t,x4,'o',label='r = 0.865')
ax[3].set_ylabel('$x_t$')
ax[3].set_xlabel('t')
ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[3].legend();
```

As expected, we see the beginnings of the period doubling route to chaos. The first r value shows one attractor, then the next one shows a period doubling where the map takes two steps to return to the same value, then the next plot shows that again where we see period four behavior, then period eight at the next r value. An important point here is the distance between the respextive r values gets much smaller every time, to the point where the period four to period 8 has a difference in r of $0.005$.

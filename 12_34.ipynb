```python
def sin_map(r,x):
  return r*sin(pi*x)
```

```python
def x_t(r):  
  x_new = [x_0]
  for i in range(t_steps):
    x_new.append(sin_map(r,x_new[i]))
  x_stable = x_new[400:500]
  return x_stable
```

```python
t_steps = 500
t_vals = 5001
x_0 = 0.1
r = linspace(0.6,1,t_vals)
```

```python
x_array = []
for r_i in r:
  x_array.append(x_t(r_i))
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.plot(r,x_array,'.',markersize = 2)
ax.set_xlabel('r')
ax.set_ylabel('x');
```

This bifurcation diagram looks very similar to the logistic map diagram, but on a different interval for r.

```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

```python
#constants and initial values 
g = 1
l = 1
omega = 0.2
R = 0.2
phi_init = 0.2
phidot_init = 0
t = np.linspace(0,20,1001)
t2 = np.linspace(0,100,1001)
t_span = [0,20]
t_span_2 = [0,100]
y0 = [phi_init, phidot_init]
```

Equation of motion $$\ddot\phi = \frac{-g}{l}\sin\phi + \frac{\omega^2 R}{l}\cos(\phi-\omega t)$$

```python
def phi_ddot(t,u):
  phi = u[0] 
  phidot = u[1]
  d = [phidot, -np.sin(phi) + omega**2*R*np.cos(phi - omega*t)]
  return d
```

```python
result_ivp = solve_ivp(phi_ddot, t_span, y0, t_eval=t)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(result_ivp.t, result_ivp.y[0])
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Phi (rads)');
```

```python
tau_natural = 2*np.pi/1
tau_natural
```

The period on this graph of the numerical solution of $\phi(t)$ is very close to that of the natural frequency of the pendulum, showing that when the $\omega$ of the hoop is small, the hoop has little effect on the oscillation of the pendulum. 

```python
result_ivp2 = solve_ivp(phi_ddot, t_span_2, y0, t_eval=t2)
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(result_ivp2.t, result_ivp2.y[0])
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Phi (rad)');
```

```python
tau = (2*np.pi)/omega
tau
```

The latter graph shows that in the long term there is a small contribution from the hoop to the oscillation of the pendulum. The amplitude of oscillations grows and shrinks periodically, with a period of about 30-35 according to the graph, which agrees with the period found from the omega of the hoop.

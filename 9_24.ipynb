```python
#IC's 
x_o = Omega = vy_1 = 1
vx_1 = vx_2 = vx_3 = vx_6 = vy_2 = 0
vy_3 = -1
vx_4 = vy_4 = -0.5
vx_5 = vy_5 = -0.7
vy_6 = -0.1
t = np.linspace(0,20,501)
```

```python
def x_t(t,x,vx,vy):
  return (x + vx*t)*np.cos(t) + (vy + x)*t*np.sin(t)
```

```python
def y_t(t,x,vx,vy):
  return -(x + vx*t)*np.sin(t) + (vy + x)*t*np.cos(t)
```

```python
fig1,ax1 = plt.subplots()
ax1.plot(x_t(t,x_o,vx_1,vy_1),y_t(t,x_o,vx_1,vy_1))
ax1.set_xlabel('x')
ax1.set_ylabel('y');
```

We have constant amplitude oscillation terms due to the initial x velocity being zero, and the second sinusoidal term's amplitude grows with time, which is why we see the outward spiral. The initial conditions are such that the position and y velocity compliment eachother, resulting in relatively large amplitudes as time grows, as $(v_{y0} + \Omega x_0)t = 2t$. 

```python
fig2,ax2 = plt.subplots()
ax2.plot(x_t(t,x_o,vx_2,vy_2),y_t(t,x_o,vx_2,vy_2))
ax2.set_xlabel('x')
ax2.set_ylabel('y');
```

We still have the constant amplitude terms, same as the first case as the initial velocity in the x dierection is still zero, however we now have zero initial y velocity, so $(v_{y0} + \Omega x_0)t = t$, which makes sense considering this case shows about half of the amplitude of the last case. 

```python
fig3,ax3 = plt.subplots()
ax3.plot(x_t(t,x_o,vx_3,vy_3),y_t(t,x_o,vx_3,vy_3))
ax3.set_xlabel('x')
ax3.set_ylabel('y');
```

In this case, the initial velocity interferes with the starting position in the time dependednt amplitude terms, as $(v_{y0} + \Omega x_0)t = 0$, so we only have the constant amplitude terms, giving us a closed orbit. 

```python
fig4,ax4 = plt.subplots()
ax4.plot(x_t(t,x_o,vx_4,vy_4),y_t(t,x_o,vx_4,vy_4))
ax4.set_xlabel('x')
ax4.set_ylabel('y');
```

In this case we have no constant amplitude terms as our x initial velocity is active. For the first term, $(x_0 + v_{x0}t)$ starts out positive, but as time progresses the initial velocity in the x term overwhelms the initial position term. For the second term in the orbut equations, since $v_{yo} < \Omega x$, the position term will eventually overwhelm the initial velocity term. The loop at the beginning represents the motion before the sign on the first term in the orbit equations flips. 

```python
fig5,ax5 = plt.subplots()
ax5.plot(x_t(t,x_o,vx_5,vy_5),y_t(t,x_o,vx_5,vy_5))
ax5.set_xlabel('x')
ax5.set_ylabel('y');
```

Same reasoning as above, however the initial velocities are slightly more negative in each direction, so the loop at the beginning of the orbit is more pronounced. 

```python
fig6,ax6 = plt.subplots()
ax6.plot(x_t(t,x_o,vx_6,vy_6),y_t(t,x_o,vx_6,vy_6))
ax6.set_xlabel('x')
ax6.set_ylabel('y');
```

We again have a constant amplitude term due to no initial velocity in the x direction, and now the initial velocity in the y counteracts the growth of the $\Omega x_0$ term, but only slows it down, as when time grows the difference between the two terms will become greater. This is why this spiral is smaller than the no initial velocity case, because the small initial velocity in the y direction makes it such that the amplitdue growth of the second terms in the orbit equations is slower.
